```python
from tensorflow.keras.utils import plot_model
from train_avsr import build_audio_model, build_video_model, build_fusion_model

import os

# Assicurati che la cartella dei risultati esista
os.makedirs('results', exist_ok=True)

audio_model = build_audio_model()
video_model = build_video_model()
fusion_model = build_fusion_model()

# Genera le immagini ad alta risoluzione
plot_model(audio_model, to_file='results/arch_audio.png', show_shapes=True, show_layer_names=True, dpi=300)
plot_model(video_model, to_file='results/arch_video.png', show_shapes=True, show_layer_names=True, dpi=300)
plot_model(fusion_model, to_file='results/arch_fusion.png', show_shapes=True, show_layer_names=True, dpi=300)
```
